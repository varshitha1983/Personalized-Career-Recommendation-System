```python
import pandas as pd
import numpy as np
from faker import Faker
import random

# Initialize Faker and parameters
fake = Faker()
num_students = 10000  # Number of rows you want
null_probability = 0.1  # Probability of a NULL value in specified columns

# skills mapped with courses
skill_courses = {
    "Statistics": ["Statistics for Data Science (edX)", "Intro to Probability and Data (Coursera)"],
    "Data Manipulation": ["Data Manipulation with pandas (Coursera)", "Data Science with R (DataCamp)"],
    "Data Visualization": ["Data Visualization with Python (Coursera)", "Data Visualization with ggplot2 (DataCamp)"],
    "Programming": ["R Programming (Coursera)", "Python for Everybody (Coursera)"],
    "Machine Learning": ["Machine Learning (Coursera by Stanford)", "Applied Data Science with Python (Coursera)"],
    "Model Evaluation": ["Model Evaluation and Improvement (Coursera)", "Statistics for Machine Learning (Udacity)"],
    "HTML": ["HTML and CSS for Beginners (Udemy)", "Introduction to HTML (Coursera)"],
    "CSS": ["CSS Fundamentals (Pluralsight)", "Advanced CSS (Udemy)"],
    "JavaScript": ["JavaScript Essentials (Udemy)", "The Complete JavaScript Course (Udemy)"],
    "Responsive Design": ["Responsive Web Design Certification (freeCodeCamp)", "Responsive Web Design Fundamentals (Coursera)"],
    "Version Control": ["Git Version Control (Udemy)", "Git & GitHub (Coursera)"],
    "Node.js": ["The Complete Node.js Developer Course (Udemy)", "Node.js for Beginners (Udemy)"],
    "Excel": ["Excel Skills for Business (Coursera)", "Mastering Excel (Udemy)"],
    "SQL": ["SQL for Data Science (Coursera)", "Learn SQL (Codecademy)"],
    "Data Cleaning": ["Data Cleaning (Coursera)", "Data Wrangling with R (DataCamp)"],
    "Data Visualization Tools": ["Data Visualization with Tableau (Coursera)", "Tableau Desktop for Beginners (Udemy)"],
    "Statistical Analysis": ["Statistics with R (Coursera)", "Statistical Inference (Coursera)"],
    "Reporting": ["Data Visualization and Reporting (Udacity)", "Storytelling with Data (Coursera)"],
    "Data Structures and Algorithms": ["Data Structures and Algorithms Specialization (Coursera)", "Algorithms, Part I (Coursera)"],
    "Front-End Frameworks": ["React - The Complete Guide (Udemy)", "Front-End Web Development with React (Coursera)"],
    "Back-End Development": ["Node.js, Express, MongoDB (Udemy)", "Django for Everybody (Coursera)"],
    "APIs": ["RESTful APIs with Flask (Udemy)", "API Development with Flask (Coursera)"],
    "Mathematics": ["Mathematics for Machine Learning (Coursera)", "Linear Algebra for Data Science (edX)"],
    "Deep Learning": ["Deep Learning Specialization (Coursera)", "Deep Learning with TensorFlow (Coursera)"],
    "Machine Learning Basics": ["Machine Learning (Coursera by Stanford)", "Intro to Machine Learning (Udacity)"],
    "Model Deployment": ["Machine Learning Engineering (Coursera)", "Deploying Machine Learning Models (Coursera)"],
    "Model Optimization": ["Hyperparameter Tuning (Coursera)", "Improving Deep Neural Networks (Coursera)"],
    "Networking Basics": ["Networking Fundamentals (Coursera)", "Computer Networking (edX)"],
    "Operating Systems": ["Linux Basics (Coursera)", "Introduction to Operating Systems (edX)"],
    "Cloud Service Providers": ["AWS Cloud Practitioner Essentials (Coursera)", "Google Cloud Fundamentals (Coursera)"],
    "Infrastructure as Code": ["Infrastructure as Code (Coursera)", "Terraform on Azure (Udemy)"],
    "Containerization": ["Docker Mastery (Udemy)", "Kubernetes for Beginners (Udemy)"],
    "Security Best Practices": ["Cloud Security Basics (Coursera)", "Introduction to Cloud Security (edX)"],
    "Cybersecurity Fundamentals": ["Cybersecurity Fundamentals (edX)", "Introduction to Cybersecurity (Coursera)"],
    "Threat Analysis": ["Threat and Vulnerability Management (Coursera)", "Introduction to Cyber Threat Intelligence (Coursera)"],
    "Incident Response": ["Incident Response and Handling (Coursera)", "Cyber Incident Response (Udacity)"],
    "Compliance Standards": ["Introduction to GDPR (Coursera)", "Compliance Frameworks (edX)"],
    "Mobile Development Frameworks": ["React Native - The Practical Guide (Udemy)", "Flutter Development Bootcamp (Udemy)"],
    "User Interface Design": ["Mobile User Experience (Coursera)", "iOS Human Interface Guidelines (Apple)"],
    "Testing": ["Mobile App Testing (Coursera)", "Automated Testing for Mobile (Udemy)"],
    "Publishing Apps": ["Publishing Android Apps (Coursera)", "App Store Optimization (Udemy)"],
    "CI/CD": ["Continuous Integration and Continuous Delivery (Coursera)", "DevOps Foundations: Continuous Integration and Continuous Delivery (LinkedIn Learning)"],
    "Scripting": ["Bash Scripting (Udemy)", "Python Scripting for Automation (Coursera)"],
    "Configuration Management": ["Ansible for Beginners (Udemy)", "Puppet Fundamentals (Udacity)"],
    "Monitoring and Logging": ["Monitoring and Logging (Coursera)", "Introduction to Logging in AWS (Udacity)"]
}

# Expanded Pathway Mapping( career goal : pathway)
pathway_mapping = {
    "Data Scientist": ("Statistics → Data Manipulation → Data Visualization → Programming → Machine Learning → Model Evaluation"),
    "Web Developer": ("HTML → CSS → JavaScript → Responsive Design → Version Control → Node.js"),
    "Data Analyst": ("Excel → SQL → Data Cleaning → Data Visualization Tools → Statistical Analysis → Reporting"),
    "Software Engineer": ("Programming Languages → Data Structures and Algorithms → Version Control → Front-End Frameworks → Back-End Development → APIs"),
    "AI Engineer": ("Mathematics → Programming → Data Preprocessing → Machine Learning Algorithms → Deep Learning → Model Evaluation"),
    "Machine Learning Engineer": ("Mathematics → Programming → Machine Learning Basics → Deep Learning → Model Deployment → Model Optimization"),
    "Cloud Engineer": ("Networking Basics → Operating Systems → Cloud Service Providers → Infrastructure as Code → Containerization → Security Best Practices"),
    "Cybersecurity Analyst": ("Networking Basics → Operating Systems → Security Principles → Threat Analysis → Incident Response → Compliance Standards"),
    "Mobile App Developer": ("Programming Languages → Mobile Development Frameworks → User Interface Design → APIs → Testing → Publishing Apps"),
    "DevOps Engineer": ("Version Control → CI/CD → Scripting → Containerization → Configuration Management → Monitoring and Logging"),
}

goal_courses = {}

for goal, skills_str in pathway_mapping.items():
    skills = skills_str.split(" → ")  # Splitting pathway string into individual skills
    courses = []
    
    for skill in skills:
        if skill in skill_courses:
            selected_course = random.choice(skill_courses[skill])  # Selecting a random course for each skill
            courses.append(selected_course)
    
    goal_courses[goal] = courses  # Assigning the list of courses to each goal


goals = ["Data Scientist", "Web Developer", "Data Analyst", "Software Engineer", "AI Engineer", "Machine Learning Engineer", "Cloud Engineer", "Cybersecurity Analyst", "Mobile App Developer", "DevOps Engineer"]
learning_styles = ["Hands-on", "Video", "Reading"]
difficulty_levels = ["Beginner", "Intermediate", "Advanced"]
course_types = ["Theory", "Practical", "Project-based"]

# Function to randomly introduce NULL values based on the specified probability
def introduce_nulls(value, column_name):
    if column_name in ["Preferred_Learning_Style", "Last_Active_Date", "Engagement_Level", "Course_Ratings"]:
        if random.random() < null_probability:
            return None
    return value

# Sample skills from the keys of skill_courses to assign to Current_Skills
def sample_skills():
    skill_keys = list(skill_courses.keys())  # Set of all skill names from skill_courses keys
    num_skills = random.randint(1, min(3, len(skill_keys)))  # Randomly choose at least 1 skill, up to 2
    return random.sample(skill_keys, k=num_skills)

career_goals = [random.choice(goals) for _ in range(num_students)]

# Define a function to calculate required skills
def calculate_required_skills(current_skills, pathway):
    if pathway is None:
        return None  # No pathway defined
    pathway_skills = set(pathway.split(" → "))  # Split pathway string into a set of skills
    current_skills_set = set(current_skills)  # Convert Current_Skills to a set
    required_skills = list(pathway_skills - current_skills_set)  # Calculate difference
    return required_skills
    
# Define a function to find the next recommended course
def get_recommended_course(goal, completed_courses):
    if goal not in goal_courses:
        return None  # No courses available for this goal
    
    goal_course_list = goal_courses[goal]  # List of all courses for the goal
    for course in goal_course_list:
        if course not in completed_courses:
            return course  # Recommend the next uncompleted course
    return "Completed Pathway"  # If all courses are completed

# Initialize course attributes for each unique course
course_attributes = {}
for courses in skill_courses.values():
    for course in courses:
        # Assign random attributes to each unique course
        course_attributes[course] = {
            "Course_Difficulty_Level": random.choice(difficulty_levels),
            "Course_Type": random.choice(course_types),
            "Course_Rating_Avg": round(random.uniform(1, 5), 1),
            "Course_Length": random.randint(5, 50),
            "Completion_Rate": round(random.uniform(50, 100), 1)
        }

# Function to get course attributes from the dictionary
def get_course_attributes(course_name):
    return course_attributes.get(course_name, {
        "Course_Difficulty_Level": None,
        "Course_Type": None,
        "Course_Rating_Avg": None,
        "Course_Length": None,
        "Completion_Rate": None
    })

# Generate dataset with relevant skills and specified NULL values
data = {
    "Student_ID": [i for i in range(1, num_students + 1)],  # Ensure Student_ID has no NULLs
    "Current_Skills": [sample_skills() for _ in range(num_students)],
    "Career_Goals": career_goals,
    "Preferred_Learning_Style": [introduce_nulls(random.choice(learning_styles), "Preferred_Learning_Style") for _ in range(num_students)],
    "Completed_Courses": [random.sample(goal_courses[goal][:3], k=2) if goal in goal_courses and len(goal_courses[goal]) >= 3 else [] for goal in career_goals],
    "Course_Ratings": [introduce_nulls(round(random.uniform(1, 5), 1), "Course_Ratings") for _ in range(num_students)],
    "Performance_Scores": [round(random.uniform(60, 100), 1) for _ in range(num_students)],
    "Engagement_Level": [introduce_nulls(random.choice(["High", "Moderate", "Low"]), "Engagement_Level") for _ in range(num_students)],
    "Weekly_Study_Hours": [random.randint(1, 20) for _ in range(num_students)],
    "Last_Active_Date": [introduce_nulls(fake.date_this_year().strftime("%Y-%m-%d"), "Last_Active_Date") for _ in range(num_students)],
    "Recommended_Course": [None] * num_students,
    "Course_Difficulty_Level": [None] * num_students,
    "Course_Type": [None] * num_students,
    "Course_Rating_Avg": [None] * num_students,
    "Course_Length": [None] * num_students,
    "Completion_Rate": [None] * num_students,
    "Required_Skills": [None] * num_students,
    "Pathway_Steps": [None] * num_students,  # Placeholder for pathway steps
}

# Create DataFrame
df = pd.DataFrame(data)

# Assign Pathway_Steps based on Career_Goals by referencing pathway_mapping
df["Pathway_Steps"] = df["Career_Goals"].apply(lambda goal: pathway_mapping[goal] if goal in pathway_mapping else None)

# Assign Required_Skills based on the difference between Pathway_Steps and Current_Skills
df["Required_Skills"] = df.apply(lambda row: calculate_required_skills(row["Current_Skills"], row["Pathway_Steps"]), axis=1)

# Assign Recommended_Course based on the next course to complete in the pathway
df["Recommended_Course"] = df.apply(lambda row: get_recommended_course(row["Career_Goals"], row["Completed_Courses"]), axis=1)

# Assign Course Attributes (Difficulty Level, Type, Rating Avg, Length, Completion Rate) for each Recommended Course
for index, row in df.iterrows():
    recommended_course = row["Recommended_Course"]
    if recommended_course and recommended_course in course_attributes:
        # Retrieve attributes from the course_attributes dictionary
        attributes = get_course_attributes(recommended_course)
        df.at[index, "Course_Difficulty_Level"] = attributes["Course_Difficulty_Level"]
        df.at[index, "Course_Type"] = attributes["Course_Type"]
        df.at[index, "Course_Rating_Avg"] = attributes["Course_Rating_Avg"]
        df.at[index, "Course_Length"] = attributes["Course_Length"]
        df.at[index, "Completion_Rate"] = attributes["Completion_Rate"]
```

```python
# Set the probability of adding noise to each column
noise_probability = 0.05  # 5% chance of noise for specified columns

# Add noise to Career_Goals by changing a small percentage to a random different career goal
def add_noise_career_goals(career_goal):
    if random.random() < noise_probability:
        return random.choice([goal for goal in goals if goal != career_goal])
    return career_goal

df["Career_Goals"] = df["Career_Goals"].apply(add_noise_career_goals)

# Add noise to Completed_Courses by adding/removing a random course
def add_noise_completed_courses(completed_courses, career_goal):
    if random.random() < noise_probability and career_goal in goal_courses:
        goal_course_list = goal_courses[career_goal]
        
        # Randomly add or remove a course
        if random.random() < 0.5 and len(completed_courses) < len(goal_course_list):
            # Add a new course from the goal pathway
            new_course = random.choice([course for course in goal_course_list if course not in completed_courses])
            completed_courses.append(new_course)
        elif completed_courses:
            # Remove a random course
            completed_courses.pop(random.randint(0, len(completed_courses) - 1))
    return completed_courses

df["Completed_Courses"] = df.apply(lambda row: add_noise_completed_courses(row["Completed_Courses"], row["Career_Goals"]), axis=1)

# Add noise to Completion_Rate by adjusting it slightly
def add_noise_completion_rate(completion_rate):
    if random.random() < noise_probability:
        # Adjust completion rate by up to ±5%
        adjustment = random.uniform(-5, 5)
        return round(min(max(completion_rate + adjustment, 50), 100), 1)  # Ensure it stays between 50 and 100
    return completion_rate

df["Completion_Rate"] = df["Completion_Rate"].apply(add_noise_completion_rate)
```

```python
# Function to simulate a slight delay or randomization in Recommended_Course
def randomize_recommendation(row):
    goal = row['Career_Goals']
    completed_courses = row['Completed_Courses']
    
    # Get the entire pathway and available courses for the goal
    pathway = pathway_mapping[goal].split(" → ") if goal in pathway_mapping else []
    goal_course_list = goal_courses[goal] if goal in goal_courses else []
    
    # Only continue if there are courses in the goal path
    if not goal_course_list:
        return row['Recommended_Course']  # Return as is if no course is available
    
    # Check if the recommended course is set and in the course list
    recommended_course = row['Recommended_Course']
    if recommended_course not in goal_course_list:
        return recommended_course  # No changes if the recommended course isn't mapped
    
    # Decide on action: 10% chance to skip the exact next course
    if random.random() < 0.1:
        # Get the index of the recommended course in goal_course_list
        course_idx = goal_course_list.index(recommended_course)
        
        # If possible, skip one course forward or backward (within bounds)
        if course_idx < len(goal_course_list) - 1:
            recommended_course = goal_course_list[course_idx + 1]
        elif course_idx > 0:
            recommended_course = goal_course_list[course_idx - 1]
    
    # 10% chance to swap with another course within the pathway
    elif random.random() < 0.1:
        # Randomly pick a different course in the pathway if it's not completed
        uncompleted_courses = [c for c in goal_course_list if c not in completed_courses]
        if uncompleted_courses:
            recommended_course = random.choice(uncompleted_courses)
    
    return recommended_course

# Apply this randomization function to each row in the DataFrame
df['Recommended_Course'] = df.apply(randomize_recommendation, axis=1)
```

```python
# Introduce some duplicate rows for duplicate handling in EDA
duplicates = df.sample(frac=0.05)  # 5% of the dataset will be duplicated
df = pd.concat([df, duplicates], ignore_index=True)

# Save to CSV and display the first few rows
df.to_csv("student_dataset.csv", index=False)
print(df.head())
```
